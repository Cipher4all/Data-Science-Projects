```
import pandas as pd
import numpy as np
iris = pd.read_csv("/content/iris.data", header=None)
```

```
iris.columns = ['Sepal length','Sepal width','Petal length','Petal Width','class labels']
```

```
x = iris[['Sepal length','Sepal width','Petal length','Petal Width']]
y = iris['class labels']
```

```
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.5, random_state=0)
```

```
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```
from sklearn import neighbors
classifier = neighbors.KNeighborsClassifier()
```

```
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
```

```
kmeans.fit(train_x, train_y)
```

```
predictions = kmeans.predict(test_x)
```

```
import seaborn as sns
sns.scatterplot(x='Petal length', y='Petal Width', hue='class labels', data=iris)
```

```
from sklearn.metrics import accuracy_score
score = accuracy_score(test_y, predictions)
print("accuracy_score:", score)
```

```
sns.pairplot(iris, hue='class labels', palette="husl", size=3)
```
